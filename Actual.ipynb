```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

# **CELLULOSE**
1200px-Cellulose_Sessel.png

```python
C = nx.Graph()
```

```python
C.add_edges_from([(1,2),
                  (2,3),
                  (3,4),
                  (3,5),
                  (5,6),
                  (6,7),
                  (7,8),
                  (8,9),
                  (7,10),
                  (10,1),
                  (5,11),
                  (11,12),
                  (12,13),
                  (13,14),
                  (14,15),
                  (13,16),
                  (16,17),
                  (17,18),
                  (18,19),
                  (18,20),
                  (20,21),
                  (17,22)])
```

```python
nx.draw(C, with_labels=True, node_color='lightblue', edge_color='grey')
```

```python
nx.wiener_index(C)
```

```python
n=C.number_of_nodes()
print(n)
sp_matrix=[]
p = dict(nx.shortest_path_length(C))
#print("do the graph has bridge:",nx.has_bridges(G))
for i in range(1,n+1):
    for j in range(1,n+1):
        
        a=p[i][j]
        sp_matrix.append(a)
sp=np.array(sp_matrix)
sp_matrix=sp.reshape(n,n)
print("shortest path \n",sp_matrix)
```

```python
upper=np.triu(sp_matrix, 1)
upper
```

```python
print("Sum of shortest path", np.sum(upper)) 
```

```python
#hyper winner index
sq_upper=np.square(upper)
sq_upper
```

```python
np.sum(sq_upper)
```

```python
hyper_winner=(np.sum(upper)+np.sum(sq_upper))/2
hyper_winner
```

```python
#first zagreb index
adj = nx.adjacency_matrix(C)
#adj
print(adj.todense())
degree=np.sum(adj, axis = 1)
print(degree)
```

```python
sq_deg=np.square(degree)
#sq_deg
```

```python
M1=np.sum(sq_deg)
M1
```
